```python
# Install (run once if needed)
# pip install transformers bertviz torch

import torch
from transformers import AutoTokenizer, AutoModel
from bertviz import head_view

# Model & tokenizer
model_name = "bert-base-uncased"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(
    model_name,
    output_attentions=True
)
model.eval()

# Input sentence
sentence = "The movie was surprisingly good and emotionally powerful."

inputs = tokenizer(sentence, return_tensors="pt")
tokens = tokenizer.convert_ids_to_tokens(inputs["input_ids"][0])

# Forward pass (no gradients)
with torch.no_grad():
    outputs = model(**inputs)

attentions = outputs.attentions  # (layers, batch, heads, seq_len, seq_len)

# Visualize attention
head_view(attentions, tokens)
```

```python
import matplotlib.pyplot as plt
import numpy as np

attention = attentions[-1][0][0].cpu().numpy()

plt.figure(figsize=(10, 8))
plt.imshow(attention)
plt.xticks(range(len(tokens)), tokens, rotation=90)
plt.yticks(range(len(tokens)), tokens)
plt.colorbar()
plt.title("Attention Heatmap (Last Layer, Head 0)")
plt.show()
```

```python
import os
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from bertviz import head_view

# Path to your local model
MODEL_DIR = os.path.abspath("model")

# Check if local model is valid
required_files = ["config.json"]
use_local = all(os.path.exists(os.path.join(MODEL_DIR, f)) for f in required_files)

if use_local:
    model_path = MODEL_DIR
else:
    print("⚠️ Local model invalid or missing — using bert-base-uncased")
    model_path = "bert-base-uncased"

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_path)

# Load model with attention enabled
model = AutoModelForSequenceClassification.from_pretrained(
    model_path,
    output_attentions=True
)

model.eval()

# Input sentence
sentence = "The movie was surprisingly good and emotionally powerful."

inputs = tokenizer(sentence, return_tensors="pt")
tokens = tokenizer.convert_ids_to_tokens(inputs["input_ids"][0])

# Forward pass
with torch.no_grad():
    outputs = model(**inputs)

attentions = outputs.attentions

# Visualize attention
head_view(attentions, tokens)

```
